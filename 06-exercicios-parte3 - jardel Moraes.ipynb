```python
alunos = [{"nome": "João", "notas": {"Matemática": 8.5, "Português": 9.0, "História": 7.0}},
          {"nome": "Maria", "notas": {"Matemática": 9.5, "Português": 8.5, "História": 6.0}},
          {"nome": "Pedro", "notas": {"Matemática": 7.0, "Português": 6.5, "História": 9.0}},
          {"nome": "Ana", "notas": {"Matemática": 8.0, "Português": 7.5, "História": 8.0}}]

def media_ponderada (alunos):
  lista = []
  
  for i in range(0,len(alunos)):
    meu_dict={}
    meu_dict['nome_p'] = alunos[i]["nome"]
    meu_dict["media"] = [((2*alunos[i]["notas"]["Matemática"]+3*alunos[i]["notas"]["Português"]+2*alunos[i]["notas"]["História"])/7)]
    lista.append(meu_dict)
  
  return lista

media_final = media_ponderada(alunos)

maior_valor = max(media_final, key=lambda x: x['media'])
print(maior_valor)

max(alunos, key=lambda alunos:[((2*alunos["notas"]["Matemática"]+3*alunos["notas"]["Português"]+2*alunos["notas"]["História"])/7)])
```

### 20. Crie um programa em Python que receba uma lista de números inteiros como entrada e calcule a soma dos quadrados dos números ímpares da lista.

```python

lista = input("Digite uma lista de números inteiros separados por espaço: ")
lista = list(map(int, lista.split()))

soma = 0

for numero in lista:
  if numero % 2 == 1:
    soma = soma + numero ** 2

print("A soma dos quadrados dos números ímpares é:", soma)
```
